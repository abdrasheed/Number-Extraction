```python
  from bs4 import BeautifulSoup
from selenium import webdriver
from time import sleep
from twilio.rest import Client
from selenium.webdriver.chrome.options import Options
from datetime import datetime

# Function to get new numbers
def get_new_numbers(numbers1, numbers2):
    new_numbers = []
    for i in numbers2:
        if i not in numbers1:
            new_numbers.append(i)
    return new_numbers


# Function to slice numbers
def chunkIt(seq):
    avg = len(seq) / 5
    out = []
    last = 0.0
    while last < len(seq):
        out.append(seq[int(last):int(last + avg)])
        last += avg
    return out


# Function to send numbers
def send_numbers(new_numbers, time_now):
    if len(new_numbers) > 170:
        sliced_numbers = chunkIt(new_numbers)
        for i in sliced_numbers:
            message_numbers = '\n,'.join(i)
            message_numbers += '\n,'+str(time_now)
            message = client.messages.create(
                body=message_numbers,
                from_='whatsapp:+14155238886',
                to='whatsapp:+96170400713'
            )
            sleep(2)

    else:
        message_numbers = '\n,'.join(new_numbers)
        message_numbers += '\n,'+str(time_now)
        message = client.messages.create(
            body=message_numbers,
            from_='whatsapp:+14155238886',
            to='whatsapp:+96170400713'
        )
url = 'https://www.touch.com.lb/autoforms/portal/touch/onlinereservation'
account_sid = 'ACf4310c704fb82e3f164bdfcdef9b01ca'
auth_token = '5175d79ffc95bf4045c9de2a0f940843'
new_numbers = []
numbers1 = []
client = Client(account_sid, auth_token)


chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('../chromedriver', chrome_options=chrome_options)
driver.get(url)

while True:
    try:
        driver.find_element_by_xpath('//*[@id=\"numbers\"]/input[9]').click()
        num1 = BeautifulSoup(driver.page_source).find('div', {'id': 'available-Numbers'}).find_all('option',{'class': 'black'})
        numbers1 = [number.text for number in num1]
    except:
        
        print('err1')
        numbers1 = []

    sleep(1)

    if len(numbers1) > 0:
        print(numbers1)
        print('sent whole messages')
        time_now = datetime.now().strftime("%H:%M:%S")
        send_numbers(numbers1, time_now)

    while len(numbers1) > 0:
        try:
            driver.find_element_by_xpath('//*[@id="numbers"]/input[9]').click()
            num2 = BeautifulSoup(driver.page_source).find('div', {'id': 'available-Numbers'}).find_all('option', {
                'class': 'black'})
            numbers2 = [number.text for number in num2]
        except:
            print('err2')
            numbers1 = []
            numbers2 = []

        # Get New Numbers
        if len(numbers2) > 0:
            new_numbers = get_new_numbers(numbers1, numbers2)

        # Send numbers
        if len(new_numbers) > 0:
            numbers1 += new_numbers
            time_now = datetime.now().strftime("%H:%M:%S")
            print(new_numbers)
            print("Current Time =", time_now)
            print("-" * 20)
            send_numbers(new_numbers, time_now)
            new_numbers = []
            sleep(1)

      
```
